```python
import math
from math import sqrt as rt
import numpy as np
from matplotlib import pyplot as plt
```

```python
x = int(input("value of x:"))
y = int(input("value of y:"))
```

```python
r = 10
```

```python
if r >= rt((x**2)+(y**2)):
    a1 = rt((r**2)-(y**2)) + x
    a2 = rt((r**2)-(x**2)) - y
    a3 = rt((r**2)-(y**2)) - x
    a4 = rt((r**2)-(x**2)) + y
else:
    a1 = 0
    a2 = 0
    a3 = 0
    a4 = 0
```

```python
print(a1, a2, a3, a4)
```

```python
# contract horizontal
b1 = 1
b3 = 1
b2 = r
b4 = r

plt.rcParams["figure.figsize"] = [15.00, 15.00]
plt.rcParams["figure.autolayout"] = True
data = np.array([[-b1, 0], [0, b2], [b3, 0], [0, -b4]])
origin = np.array([[0, 0, 0, 0], [0, 0, 0, 0]])
plt.quiver(*origin, data[:, 0], data[:, 1], color=['black', 'red', 'green', 'yellow'], scale=30)
plt.show()
```

```python
# adjust vertical
b1 = 1
b3 = 1
b2 = y
b4 = 2*r-y

plt.rcParams["figure.figsize"] = [15.00, 15.00]
plt.rcParams["figure.autolayout"] = True
data = np.array([[-b1, 0], [0, b2], [b3, 0], [0, -b4]])
origin = np.array([[0, 0, 0, 0], [0, 0, 0, 0]])
plt.quiver(*origin, data[:, 0], data[:, 1], color=['black', 'red', 'green', 'yellow'], scale=30)
plt.show()
```
